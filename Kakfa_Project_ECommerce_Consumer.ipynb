```python
def ProcessingMessage(message):
  try:
    if message.error():
      if (message.error.code()== KafkaError._PARTOTION_EOF):
        print(f'End of partition reached {0}/ {1}')

      else:
        raise KafkaException(message.error)

    else:
      key= message.key().decode('utf-8')
      value= message.value().decode('utf-8')
      print(f'Message received key: {key}, Value: {value}')

  except Exception as e:
    print(f'Exception is {e}')


try:
  print('Listening for message Press CTRL+C to exit')
  while True:
    msg= consumer.poll(timeout=1.0)
    if msg:
      ProcessingMessage(msg)

except KeyboardInterrupt:
  print("Interrupted by user, Shutting down Consumer")
finally:
  consumer.close()




```
