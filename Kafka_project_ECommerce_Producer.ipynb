```python
topic= 'E-Commerce'
with open(json_file, 'r') as file:
  california_housing_traing_data= json.load(file)


```

```python
value= california_housing_traing_data[0]
key= value['longitude']
producer.produce(topic, key= str(key).encode('utf-8') , value= str(value).encode('utf-8'))
producer.flush()

#
```

```python
def delivery_status(err,msg):
  if err:
    print('Message Failed to Deliver')
  else:
    print(f'Message delivered to {msg.topic} {msg.partition()} at offset {msg.offset()}')


for record in california_housing_traing_data:
      message_value= json.dumps(record)
      message_key= str(record['longitude']).encode('utf-8')
      producer.produce(topic,key= message_key, value= message_value, callback= delivery_status)

producer.flush()
```
